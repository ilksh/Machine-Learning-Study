### example

#### 1. Extracting only data from the continents of Asia and Europe from the city_data
- city_data.loc[city_data["Continent" == "Asia" or "Europe]

#### 2. Making data that is only 'Europe' and 'Asia' continent by combining the extracted data
- new_data = pd.concat(asia, europe)
 
#### 3. Output of the number of cities in Europe and Asia, respectively
- for (continent, group) in grouped:

#### 4. Designation of Europe as blue and Asia as red
- "Asia" : "red"
- "Europe" : "blue"

#### 5. Drawing a scatter plot to x-axis Planet and y-axis Profit from Europe and Asia
- plt.scatter(x_value, y_value, c = color)

```python
city_data.head()
```

```python
# Step 1
# Extracting only data from the continents of Asia and Europe from the city_data

asia_group = city_data.loc[city_data["Continent"] == "Asia"]
europe_group = city_data.loc[city_data["Continent"] == "Europe"]
```

```python
asia_group.head()
```

```python
europe_group.head()
```

```python
# Step 2
# Making data that is only 'Europe' and 'Asia' continent by combining the extracted data

asia_europe = pd.concat([asia_group, europe_group])

asia_europe
```

```python
grouped = asia_europe.groupby("Continent")

grouped.groups
```

```python
# Step 3
# Output of the number of cities in Europe and Asia, respectively

for continent, group in grouped:
    print("{} : {}".format(continent, len(group)))
```

```python
# Step 4
# Designation of Europe as blue and Asia as red

continent_color = {"Asia": 'red', 'Europe': 'blue'}

# Step 5
# Drawing a scatter plot to x-axis Planet and y-axis Profit from Europe and Asia

plt.figure(figsize=(13, 7))

for idx, item in grouped:
    if idx == 'Asia' or idx == 'Europe':
        plt.scatter('Planet', 'Profit', data = item, label = idx, c = continent_color[idx])

plt.xlabel('Planet')
plt.ylabel('Profit')
plt.legend()
plt.show()
```
