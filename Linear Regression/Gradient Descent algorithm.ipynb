### 1. Logic

#### Gradient Descent algorithm
- By using grdient descent algorithm, move wo, w1 to the minimum value
- wo = wo + (learning rate) * delta(w0)
- w1 - w1 + (learning rate) * delta(w1)


### 2. Optimization
- Iterate 'echo' times, training data to find the optimal value
- update w0, w1 value

```python
# data set
x = [130, 165, 150, 150, 140, 198, 220, 215, 225, 190, 170, 160, 150]
y = [3504, 3693, 3436, 3433, 3449, 4341, 4354, 4312, 4425, 3850, 3563, 3609, 3761]

plt.title('Sccater plot')
plt.xlabel('x')
plt.ylabel('y')
plt.scatter(x,y)
```

```python
# number of data
m = len(x)

# h(x) = w0 + w1*x
# initialize value
w0 = 0
w1 = 1

# number of repetition of learning
echo = 100000 

# learning rate
alpha = 0.00002

# Gradient descent algorithm
for _ in range(echo):
    
    # w0_step = partial differentials of cost function to w0
    w0_step = sum([w0 + w1*x_i - y_i for x_i, y_i in zip(x,y)])
    
    # w1_step = partial differentials of cost function to w1
    w1_step = sum([(w0 + w1*x_i - y_i) * x_i for x_i, y_i in zip(x,y)])
    
    aa = alpha * w0_step
    bb = alpha * w1_step
    
    # backup data
    temp0 = w0 - aa/m
    temp1 = w1 - bb/m
    
    # data update
    wo = temp0
    w1 = temp1
    
x_range = np.linspace(min(x), max(x))

plt.title('Sccater plot')
plt.xlabel('x')
plt.ylabel('y')
plt.scatter(x,y)

plt.plot(x_range, w0 + w1*x_range, color='red')

plt.show()
```
