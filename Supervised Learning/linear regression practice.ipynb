```python
data = pd.read_csv("Data Set/CarPrice_Assignment.csv")
data
```

```python
df = pd.DataFrame(data)
df = df[['horsepower', 'peakrpm','price']]
df
```

```python
plt.figure(figsize = (10,10))
plt.scatter(df.horsepower, df.price)
```

```python
m = len(df)
w0 = 0
w1 = -1
echo = 5000
learning_rate = 0.00002
```

```python
for iter in range(echo):
  aa = learning_rate*sum([w0 + w1*x - y for x, y in zip(df.horsepower, df.price)])
  bb = learning_rate*sum([(w0 + w1*x - y)*x for x, y in zip(df.horsepower, df.price)])

  temp0 = w0 - aa / m
  temp1 = w1 - bb / m

  w0 = temp0
  w1 = temp1
```

```python
plt.figure(figsize = (10,10))
x_range = np.linspace(min(df.horsepower), max(df.horsepower))
y_range = w0 + w1*x_range
plt.scatter(df.horsepower, df.price)
plt.plot(x_range, y_range, '-r')

plt.show()
```
