```python
m = len(df)
w = pd.Series([np.random.rand(),np.random.rand()])

echo = int(1e4)   # number of repitition
learning_rate = 0.0002  

print(w)
```

```python
x = df.age
new_x = [[1 for _ in range(m)], df.age]
y = df.glucose

for _ in range(echo):
    w_dot = w.dot(new_x)
    w_dot -= y
    
    delta_w0 = sum(w_dot) / m
    delta_w1 = w_dot.dot(x) / m
    
    w[0] = w[0] - learning_rate * delta_w0
    w[1] = w[1] - learning_rate * delta_w1
    
```
